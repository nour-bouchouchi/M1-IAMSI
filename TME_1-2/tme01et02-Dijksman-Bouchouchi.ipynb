### Structure de données

Pour représenter une position d'Awélé en Python. Une *position* est définie par les éléments suivants :
- une *dimension* qui correspond au nombre de colonnes (dans l'Awélé original, il y a 6 colonnes). On note $n$ ce nombre de colonnes dans la suite de ce texte ;
- un *plateau* qui correspond au plateau de jeu et donne le nombre de graines dans chaque case ;
- un *nombre* qui correspond au nombre total de graines sur le plateau au début du jeu (dépend du nombre de dimensions et du nombre de graines dans chaque case au départ) ;
- le *trait* qui correspond au camp devant jouer dans la position donnée. Il y a
  deux camps: "NORD et "SUD" ;
- le *gain* qui correspond au nombre de graines déjà ramassées par chacun des joueurs.

En Python, on choisit de représenter une telle position sous la forme d'un dictionnaire qui associe à chaque nom de champ sa valeur.

        # Représentation Python d'une position du jeu d'Awélé:
        position = dict()       # initialisation
        position['plateau']     # les cases du jeu (voir ci-dessous)
        position['dimension']   # le nombre de colonnes du plateau
        position['nb_total']    # nombre de graines total au début du jeu sur tout le plateau
        position['trait']       # le joueur qui doit jouer: 'SUD' ou 'NORD'
        position['gain']        # les graines prises par chaque joueur sous la forme : {'SUD':0, 'NORD':0}


Dans ce qui suit, on distingue :
- un coup *correct* qui correspond à un numéro de colonne correct (valeur comprise entre 1 et $n$), et pour lequel la case correspondante contient au moins une graine;
- un coup *autorisé* qui est un coup à la fois correct et permettant d'atteindre une position légale du jeu, c'est-à-dire une position qui respecte la règle obligeant à laisser au moins une graine dans le camp de son adversaire.

### Représentation du plateau de jeu en Python 


Pour représenter le plateau en Python, on utilise une liste d'entiers, chaque entier correspondant au nombre de graines dans une case. 

Soit un plateau de jeu de $n$ colonnes.

Pour le camp SUD :
- le nombre de graines de la case de la colonne 1 est stocké en position $0$ dans la liste Python ;
- le nombre de graines de la case de la colonne  2 est stocké en position $1$ dans la liste Python ;
- ...
- le nombre de graines de la case de la colonne  $n$ est stocké en position $n-1$ dans la liste Python.

Pour le camp NORD :
- le nombre de graines de la case de la colonne  1 est stocké en position $2n-1$ dans la liste Python ;
- le nombre de graines de la case de la colonne  2 est stocké en position $2n-2$ dans la liste Python ;
- ...
- le nombre de graines de la case de la colonne  $n$ est stocké en position $n$ dans la liste Python.

Cette représentation à l'avantage de favoriser le parcours des cases du plateau : en parcourant la liste par positions croissantes, on retrouve le parcours en sens inverse des aiguilles d'une montre. Ce parcours peut alors se faire en Python en incrémentant de 1 la position pour passer à la case suivante, et en réalisant un calcul modulo $2n$.

<font color="RED" size="+1">**[Q]**</font> Ecrire les fonctions suivantes :
- <tt>est_correct(position,nombre)</tt>: rend le booléen <tt>True</tt> si le nombre donné peut être un coup correct dans la position donnée, c'est-à-dire si ce nombre est un entier compris entre 1
  et <tt>n</tt> et si de plus la case correspondante du plateau (pour le camp qui doit donc jouer
  dans la position) contient au moins une graine.
- <tt>est_legale(position)</tt>: rend le booléen <tt>True</tt> si la position donnée est légale, c'est-à-dire si le camp qui doit jouer possède au moins une dans son camp.
- <tt>effectue_si_valide(position,coup)</tt>: rend la nouvelle position obtenue en jouant <tt>coup</tt>
  dans la position donnée. Cette fonction retourne le booléen <tt>False</tt> si le coup
  n'est pas correct ou si la position résultante n'est pas une position légale du jeu.
- <tt>est_terminale(position)</tt>: rend le booléen <tt>True</tt> si la position est
  terminale, c'est-à-dire si aucun coup n'est correct, ou si tout coup correct ne permet
  pas d'atteindre une position légale ou bien si l'un des deux joueurs a gagné un nombre de graines suffisant. Dans le jeu Awélé original à 6 colonnes, la position est gagnée dès qu'un camp a remporté 25 graines. Ce qui correspond à $6*4+1$ graines. Ce nombre de graines nécessaire pour remporter la victoire dépend donc de la taille du plateau de jeu.


```python
def est_correct(position,nombre):
    """
    Retourne True si le nombre donné peut être un coup correct dans la position donnée et False sinon.
    """
    n = position['dimension'] #nombre de cases dans chaque camp sur le plateau
    #Si le nombre n'est pas compris entre 1 et n-1 le coup est incorrect
    if nombre>n or nombre<1: 
        return False
    #Le joueur SUD a les cases d'indice 0 à n-1 : la case nombre a l'indice nombre-1
    if position['trait'] == 'SUD':
        return position['plateau'][nombre-1]>0
    #Le joueur NORD a les cases d'indice 2n-1 à n : la case nombre a l'indice 2n-nombre
    else:
        return position['plateau'][2*n-nombre]>0

    
def est_legale(position):
    """
    Retourne True si la position est légale et False sinon.
    """
    n = position['dimension']
    #Si c'est à SUD de jouer il faut qu'il y ait au moins une graine dans son camp
    if position['trait'] == 'SUD':
        return sum(position['plateau'][0:n])>0
    #De même si c'est à NORD de jouer
    else:
        return sum(position['plateau'][n:2*n])>0

    
def effectue_si_valide(position,coup):
    """
    Retourne la position obtenue suite au coup si la position obtenue est légale.
    Retourne False si le coup n'est pas correct ou que la position obtenue n'est pas légale.
    """
    position2 = aw.duplique(position) #on duplique la position pour ne pas la modifier
    position2 = aw.joue_un_coup(position,coup) #et on joue le coup
    
    #Si le coup n'est pas correct ou que la position obtenue n'est pas légale on retourne False...
    if not(est_correct(position,coup)) or not(est_legale(position2)):
        #print("Coup non valide")
        return False
    #... sinon on retourne la position obtenue suite au coup
    else:
        return position2

    
def exists_winner(position):
    """
    Retourne True si un des joueurs a gagné suffisemment de graines et False sinon.
    """
    nb_graines = position['nb_total']/2 #nombre de graines par camp au début du jeu
    win = nb_graines + 1 #nombre de graines qu'un joueur doit récupérer pour gagner 
    
    if position['gain']['SUD'] >= win or position['gain']['NORD'] >= win : 
        return True
    
    return False


def est_terminale(position):
    """
    Retourne True si un joueur a gagné ou qu'aucun coup ne peut être joué.
    """
    n = position['dimension'] #nombre de cases dans chaque camp sur le tablier

    #Si SUD ou NORD a gagné suffisemment de graines : retourne True
    if exists_winner(position):
        return True
    
    #Sinon on retourne True si aucun coup ne peut être joué
    for coup in range(1,n+1):
        if not(effectue_si_valide(position,coup)==False):
            return False
    return True
```

### Moteur de jeu simple

<font color="RED" size="+1">**[Q]**</font> Ecrire la fonction <tt>partie_humains(taille=6,nb_graines=4)</tt> permettant à deux joueurs humains de s'affronter dans une partie d'Awélé sur un plateau qui contient <tt>taille</tt> colonnes et <tt>nb_graines</tt> par case au départ.  
  Dans cette fonction, les joueurs entrent à tour de rôle leur coup à jouer. Après chaque coup, la position est affichée avec les gains de chaque joueur. Le jeu s'arrête lorsque une position terminale est atteinte et le programme affiche alors le nom du camp vainqueur.
  Durant la partie, les coups sont donnés en utilisant la fonction <tt>input()</tt> de Python.
  


```python
def partie_humains(taille = 6, nb_graines = 4):
    """
    Permet à deux personnes de s'affronter dans une partie d'Awélé sur un plateau de taille 'taille' 
    avec au début de la partie 'nb_graines' graines par case.
    """
    #Initialisation du plateau et afficage de celui-ci
    position = aw.position_initiale(taille,nb_graines)
    aw.affichage(position)
    
    #Tant qu'il est possible de jouer...
    while not(est_terminale(position)):
        
        #On récupère le coup de la personne qui joue
        coup = int(input("Entrez votre coup : "))
        #Tant que le coup entré n'est pas valide on redemande au joueur d'entrer un coup valide
        while effectue_si_valide(position,coup)==False:
            coup = int(input("Veuillez entrer un coup valide : "))
        
        #On met à jour la position
        position = aw.joue_un_coup(position,coup)
                
        #On affiche le plateau
        aw.affichage(position)
        
    #S'il n'est plus possible de jouer parce qu'aucun coup ne peut être joué
    #l'adversaire de celui qui ne peut pas jouer récupère toutes les graines
    if not(exists_winner(position)):
        if position['trait'] == 'SUD':
            print("Aucun coup n'est possible : NORD récupère toutes les graines sur le tablier")
            position['gain']['NORD'] = position['nb_total'] - position['gain']['SUD']
        else:
            print("Aucun coup n'est possible : SUD récupère toutes les graines sur le tablier")
            position['gain']['SUD'] = position['nb_total'] - position['gain']['NORD']
        for i in range(len(position['plateau'])):
            position['plateau'][i]=0
        aw.affichage(position)

    #Lorsque la position est terminale soit un joueur a remporté suffisemment de graines pour gagner la partie
    #(il s'agit alors du joueur ayant le plus de graines), soit aucun coup ne pouvait être joué, le vainqueur
    #est alors le joueur avec le plus de graines. 
    if position['gain']['SUD'] > position['gain']['NORD']:
        print("Le camp SUD a gagné")
    else:
        if position['gain']['SUD'] < position['gain']['NORD']:   
            print("Le camp NORD a gagné")
        else :
            print("Il y a égalité")
```

<font color="RED" size="+1">**[Q]**</font> Ecrire la fonction <tt>genere_un_coup(position)</tt> qui retourne un coup autorisé pour la position donnée. Le coup est choisi de façon aléatoire parmi les coups possibles dans la position donnée. Cette fonction rend la valeur 0 si aucun coup n'est possible.

```python
import random

def coups_possibles(position):
    """
    Retourne la liste des coups posssibles.
    """
    n = position['dimension']
    liste_coups_poss = []
  
    #Pour chaque coup possible 
    for coup in range(1,n+1):
        #On vérfie en premier s'il y a bien des graines dans la case
        if est_correct(position, coup) : 
            position2 = aw.duplique(position) #on duplique la position pour ne pas la modifier
            position2 = aw.joue_un_coup(position,coup) #et on joue le coup
            #Si le coup permet d'arriver dans une position légale...
            if est_legale(position2):
                liste_coups_poss.append(coup) #on ajoute le coup à la liste des coups possibles
    return liste_coups_poss

def genere_un_coup(position):
    """
    Retourne un coup autorisé pour la position donné (choix aléatoire).
    Retourne 0 si aucun coup n'est possible.
    """

    liste_coups_poss = coups_possibles(position) 

    if liste_coups_poss==[]:
        return 0
    
    else:
        return liste_coups_poss[random.randint(0,len(liste_coups_poss)-1)]
```

## Joueur artificiel simple

On considère la fonction d'évaluation suivante de la position $P$ :
$$f(P) = \left.
  \begin{array}{ll}
        +99      & \mbox{si } P \mbox{ est gagnée par sud} \\
        -99      & \mbox{si } P \mbox{ est gagnée par nord} \\
        (2f_g(P,sud) + f_{12}(P,nord)) - (2f_g(P,nord) + f_{12}(P,sud))& \mbox{sinon }
   \end{array} \right.$$

avec
- $f_g(P,c)$ : nombre de graines déjà gagnées par le camp $c$ dans la position $P$
- $f_{12}(P,c)$ : nombre de cases du camp $c$ contenant 1 ou 2 graines, dans la position $P$.

```python
def f12(P, camp):
    """
    Renvoie le nombre de cases du camp contenant 1 ou 2 graines dans la position P
    """
    n = P['dimension']
    #Pour le camp SUD
    if camp=="SUD" : 
        #Retourne la longueur de la liste après filtrage (ne conserve que les cases avec une ou deux graines)
        return len( [i for i in P['plateau'][:n] if (i==1 or i==2)] )
    #Idem pour le camp NORD
    else :
        return len( [i for i in P['plateau'][n:] if (i==1 or i==2)] )

def evalue(position):
    """
    Renvoie la valeur de l'évaluation de la position obtenue en utilisant la fonction f
    """
    nb_graines = position['nb_total']/2 #nombre de graines par camp au début du jeu
    win = nb_graines + 1 #nombre de graines qu'un joueur doit récupérer pour gagner

    #Si position est gagnée par SUD
    if position['gain']["SUD"]>=win : 
        return 99

    #Si position est gagnée par NORD
    if position['gain']["NORD"]>=win : 
        return -99
  
    #Sinon
    return (2*position['gain']['SUD'] + f12(position,'NORD'))-(2*position['gain']['NORD'] + f12(position, 'SUD'))
```

```python
def evalueMin(position, prof):
    """
    Evaluation lorsque le but est de minimiser le gain (on se trouve au niveau d'un noeud MIN)
    """
    #cas de base (on se trouve au niveau d'une feuille ou on ne peut plus jouer car la position est terminale)
    if prof==0 or est_terminale(position):
        return evalue(position)

    val_meilleur_coup = sys.maxsize #le but sera de minimiser cette valeur

    #On cherche le coup (parmi tous les coups possibles depuis la position actuelle) le plus avantageux
    for coup in coups_possibles(position):
        position2 = aw.duplique(position)
        position2 = aw.joue_un_coup(position,coup)
        val_meilleur_coup2 = evalueMax(position2,prof-1)
        if val_meilleur_coup2 < val_meilleur_coup : #on cherche à minimiser le gain
            val_meilleur_coup = val_meilleur_coup2
  
    return val_meilleur_coup


def evalueMax(position, prof):
    """
    Evaluation lorsque le but est de maximiser le gain (on se trouve au niveau d'un noeud MAX)
    """
    #cas de base (on se trouve au niveau d'une feuille ou on ne peut plus jouer car la position est terminale)
    if prof==0 or est_terminale(position):
        return evalue(position)

    val_meilleur_coup = -sys.maxsize #le but sera de maximiser cette valeur

    #On cherche le coup (parmi tous les coups possibles depuis la position actuelle) le plus avantageux
    for coup in coups_possibles(position):
        position2 = aw.duplique(position)
        position2 = aw.joue_un_coup(position,coup)
        val_meilleur_coup2 = evalueMin(position2,prof-1)
        if val_meilleur_coup2 > val_meilleur_coup : #on cherche à maximiser le gain
            val_meilleur_coup = val_meilleur_coup2
  
    return val_meilleur_coup


def minimax(position, prof):
    """
    Retourne un tuple dont : 
    - le premier terme est le meilleur coup à jouer pour une profondeur prof
    - le second terme est l'évaluation de la position obtenue pour ce coup
    """
    #Si la joeur est SUD on se trouve sur une case MAX (initialise à - l'infinfi)
    if(position['trait']=='SUD'):
        val_meilleur_coup = -sys.maxsize 
    #SI c'est NORD on se trouve sur une case MIN (initialise à + l'infini)
    else : 
        val_meilleur_coup = sys.maxsize 

    meilleur_coup = -1
    
    #On teste tous les coups possibles depuis la position de départ en dupliquant cette position et en jouant le coup
    for coup in coups_possibles(position):
        position2 = aw.duplique(position)
        position2 = aw.joue_un_coup(position,coup)
        
        #Si le joueur est SUD on se trouve au début sur un noeud MAX
        if position['trait']=='SUD':
            val_meilleur_coup2 = evalueMin(position2,prof-1)
            #Si on a trouvé un meilleur coup on modifie le meilleur coup et sa valeur
            if val_meilleur_coup2 > val_meilleur_coup :
                val_meilleur_coup = val_meilleur_coup2
                meilleur_coup = coup
                
        #Si le joueur est NORD on se trouve au début sur un noeud MIN
        else:
            val_meilleur_coup2 = evalueMax(position2,prof-1)
            #Si on a trouvé un meilleur coup on modifie le meilleur coup et sa valeur
            if val_meilleur_coup2 < val_meilleur_coup :
                val_meilleur_coup = val_meilleur_coup2
                meilleur_coup = coup
  
    #On renvoie le meilleur coup et sa valeur
    return meilleur_coup, val_meilleur_coup
```

```python
def partie_minimax(campCPU, taille=6, nb_graines=4, prof=2):
    """
    Permet de jouer une partie entre un joueur humain et un ordinateur utilisant l'algorithme minimax avec une profondeur prof
    """
    #Si campCPU ne vaut pas 'NORD' ou 'SUD':
    if not(campCPU=='NORD' or campCPU=='SUD'):
        print("Il n'existe que deux camps possibles : 'NORD' et 'SUD'")
        return -1

    #Initialisation du plateau et afficage de celui-ci
    position = aw.position_initiale(taille,nb_graines)
    aw.affichage(position)
  
    #Tant qu'il est possible de jouer...
    while not(est_terminale(position)):
        #Si c'est à l'ordinateur de jouer :
        if position['trait']==campCPU : 
            coup,valeur = minimax(position,prof) #Le coup généré est forcément valide...
            position = aw.joue_un_coup(position,coup) #...il n'est donc pas utile de vérifier si le coup est valide

        #Sinon (c'est au joueur humain de jouer)
        else :
            #On récupère le coup de la personne qui joue
            coup = int(input("Entrez votre coup : "))
            #Tant que le coup entré n'est pas valide on redemande au joueur d'entrer un coup valide
            while effectue_si_valide(position,coup)==False:
                coup = int(input("Veuillez entrer un coup valide : "))
        
            #On met à jour la position
            position = aw.joue_un_coup(position,coup)
                
        #On affiche le plateau
        aw.affichage(position)
      
    #S'il n'est plus possible de jouer parce qu'aucun coup ne peut être joué
    #l'adversaire de celui qui ne peut pas jouer récupère toutes les graines
    if not(exists_winner(position)):
        if position['trait'] == 'SUD':
            print("Aucun coup n'est possible : NORD récupère toutes les graines sur le tablier")
            position['gain']['NORD'] = position['nb_total'] - position['gain']['SUD']
        else:
            print("Aucun coup n'est possible : SUD récupère toutes les graines sur le tablier")
            position['gain']['SUD'] = position['nb_total'] - position['gain']['NORD']
        for i in range(len(position)):
            position['plateau'][i]=0
        aw.affichage(position)

    #Lorsque la position est terminale soit un joueur a remporté suffisamment de graines pour gagner la partie
    #(il s'agit alors du joueur ayant le plus de graines), soit aucun coup ne pouvait être joué, le vainqueur
    #est alors le joueur avec le plus de graines. 
    if position['gain']['SUD'] > position['gain']['NORD']:
        print("Le camp SUD a gagné")
    else:
        if position['gain']['SUD'] < position['gain']['NORD']:   
            print("Le camp NORD a gagné")
        else :
            print("Il y a égalité")
```

```python
def min_alphabeta(position, prof, alpha, beta):
    """
    Retourne le beta obtenu dans le noeud MIN
    """
    #cas de base 
    if prof==0 or est_terminale(position):
        return evalue(position)

    liste_coups_possibles = coups_possibles(position) #liste des coups possibles
    i = 0
    #pour chaque coup possible (tant que alpha<beta) on simule le coup et on l'évalue
    while( i<len(liste_coups_possibles) and alpha<beta ) : 
        position2 = aw.duplique(position)
        position2 = aw.joue_un_coup(position,liste_coups_possibles[i])
        beta2 = max_alphabeta(position2,prof-1,alpha,beta)
        if(beta>beta2):
            beta = beta2
        i += 1
    return beta


def max_alphabeta(position, prof, alpha, beta):
    """
    Retourne le alpha obtenu dans le noeud MAX
    """
    #cas de base 
    if prof==0 or est_terminale(position):
        return evalue(position)
    
    liste_coups_possibles = coups_possibles(position) #liste des coups possibles
    i = 0
    #pour chaque coup possible (tant que alpha<beta) on simule le coup et on l'évalue
    while( i<len(liste_coups_possibles) and alpha<beta ) : 
        position2 = aw.duplique(position)
        position2 = aw.joue_un_coup(position,liste_coups_possibles[i])
        alpha2 = min_alphabeta(position2,prof-1,alpha,beta)
        if(alpha<alpha2):
            alpha = alpha2
        i += 1
    return alpha


def alphabeta(position,prof=2,alpha=-sys.maxsize,beta=sys.maxsize):
    """
    Retourne un tuple dont : 
    - le premier élément est le meilleur coup à jouer selon l'algorithme alpha-beta
    - le second élément est la valeur de ce coup
    """
    #cas de base 
    if prof==0 or est_terminale(position):
        return evalue(position)
    
    best_coup = 0
    liste_coups_possibles = coups_possibles(position)
    i = 0
    
    #Pour tous les coups possibles et tant que alpha < beta (sinon coupe)
    while( i<len(liste_coups_possibles) and alpha<beta ) : 
        position2 = aw.duplique(position)
        position2 = aw.joue_un_coup(position,liste_coups_possibles[i])
        
        #Si le joueur est SUD on se trouve au dévut sur un noeud MAX
        if(position['trait']=='SUD'):
            alpha2 = min_alphabeta(position2,prof-1,alpha,beta) #transmet alpha et beta au noeud min et récupère le nouveau alpha
            if(alpha<alpha2): #si le alpha est plus grand (meilleur gain)
                alpha = alpha2
                best_coup = liste_coups_possibles[i]
                
        #Si le joueur est NORD on se trouve au début sur un noeud MIN 
        else : 
            beta2 = max_alphabeta(position2,prof-1,alpha,beta)
            if(beta>beta2):
                beta = beta2
                best_coup = liste_coups_possibles[i]
        
        i += 1
  
    return best_coup, alpha
```

```python
class Partie2J : 
    def __init__(self, taille, nb_graines):
        self.position = aw.position_initiale(taille,nb_graines)
        self.fenetre = 0
        self.N_S_CLIC = -1 #permet de savoir si le clic a été effectué dans la partie NORD ou SUD du plateau
        self.NUM_CASE_CLIC = -1 #permet de savoir sur quelle case a été effectué le clic
        self.fin = False
    
            
    def draw_plateau(self):
        """
        Permet de mettre à jour la position du plateau
        """
        nb_cases = self.position["dimension"] #nombre de case par camp
        dessin = tk.Canvas(self.fenetre, width=100*nb_cases+20, height=500 , bg='white')
        dessin.grid(row = 0, column = 0, columnspan=2, padx=3, pady=3)
        lignes = []
        lignes.append(dessin.create_line(10, 100, 100*nb_cases+10, 100, width=3))
        lignes.append(dessin.create_line(10, 200, 100*nb_cases+10, 200, width=3))
        lignes.append(dessin.create_line(10, 300, 100*nb_cases+10, 300, width=3))
        for i in range(nb_cases+1):
            lignes.append(dessin.create_line(10+100*i, 100, 10+100*i, 300, width=3))
        nb_cases = self.position["dimension"] #nombre de case par camp
        tk.Label(text="Nord : "+str(self.position['gain']['NORD'])+" prises", font=('Helvetica', 14), bg = "white").place(x = 15, y = 50)
        tk.Label(text="Sud : "+str(self.position['gain']['SUD'])+" prises", font=('Helvetica', 14), bg = "white").place(x = 15, y = 350)
        for i in range(nb_cases):
            tk.Label(text=str(self.position['plateau'][i]), font=('Helvetica', 20), bg = "white").place(x = 50+i*100, y = 240)
        for i in range(nb_cases, 2*nb_cases):
            tk.Label(text=str(self.position['plateau'][i]), font=('Helvetica', 20), bg = "white").place(x = 60+(nb_cases-1-(i%nb_cases))*100, y = 140)
        
        if not(self.fin) : 
            tk.Label(text="C'est au tour de : "+str(self.position['trait']), font=('Helvetica', 14), bg = "white").place(x = 15, y = 420)        

    
    def clic(self,event):
        """ 
        Récupère la position du clic et retourne la position de la case sur laquelle on a cliqué 
        ou -1 si on a cliqué en dehors du plateau (le numéro et s'il s'agit d'une case NORD ou SUD)
        """
        # position du pointeur de la souris
        X = event.x
        Y = event.y
        nb_cases = self.position["dimension"] #nombre de case par camp
        if(Y>=100 and Y<=200):
            self.N_S_CLIC = 'NORD'
        else : 
            if(Y>=200 and Y<=300):
                self.N_S_CLIC = 'SUD'
            else : 
                self.N_S_CLIC = -1
        if(not(X<10) and not(X>10+100*nb_cases)):
            self.NUM_CASE_CLIC = X//110+1
        
    def joue_coup_valide(self):
        """
        Permet de jouer un coup si celui-ci est valide.
        On rappelle la fonction tant que ce n'est pas le cas.
        """
        n_s = self.N_S_CLIC
        coup = self.NUM_CASE_CLIC
        #Si le coup est valide on joue le coup
        if n_s==self.position['trait'] and not(effectue_si_valide(self.position,coup)==False):
            self.position = aw.joue_un_coup(self.position,coup)
            return
        #Sinon on rappelle la fonction 
        self.fenetre.after(10, self.joue_coup_valide)
    
    def fin_partie(self):
        """
        Si la partie est finie : distinction du cas où les joueurs ne peuvent plus jouer ee
        celui où un joueur a récolté suffisamment de graines.
        """
        if not(exists_winner(self.position)):
            mess = ""
            if self.position['trait'] == 'SUD':
                mess = "Aucun coup n'est possible : NORD récupère toutes les graines sur le tablier"
                self.position['gain']['NORD'] = self.position['nb_total'] - self.position['gain']['SUD']
            else:
                mess = "Aucun coup n'est possible : SUD récupère toutes les graines sur le tablier"
                self.position['gain']['SUD'] = self.position['nb_total'] - self.position['gain']['NORD']
            for i in range(len(self.position['plateau'])):
                self.position['plateau'][i]=0
            self.draw_plateau()
            tk.Label(text=mess, font=('Helvetica', 14), bg = "white").place(x = 15, y = 420)        

                
        if self.position['gain']['SUD'] > self.position['gain']['NORD']:
            tk.Label(text="Le camp SUD a gagné", font=('Helvetica', 14), bg = "white").place(x = 15, y = 460)        
        else:
            if self.position['gain']['SUD'] < self.position['gain']['NORD']:   
                tk.Label(text="Le camp NORD a gagné", font=('Helvetica', 14), bg = "white").place(x = 15, y = 460)        
            else :
                tk.Label(text="Il y a égalité", font=('Helvetica', 14), bg = "white").place(x = 15, y = 460)        
        return
        
    def jeu(self):
        """
        Simule une partie (continue tant que le jeu n'est pas fini).
        """
        #On commence par jouer un coup valide
        self.joue_coup_valide()
        self.draw_plateau()
        #Une fois qu'un coup a été joué on regarde si la position est terminale
        if(est_terminale(self.position)):
            self.fin=True
            self.fin_partie() #si c'est le cas la partie est finie
            return
        self.fenetre.after(100, self.jeu) #sinon on refait un tour de jeu (rappelle la fonction jeu)

            
    def partie(self):
        """
        Permet de créer la fenêtre graphique (afficher le plateau) et de lancer le jeu.
        """
        self.fenetre = tk.Tk()
        self.fenetre.title("Awélé 2 joueurs")
        self.draw_plateau()
        self.fenetre.bind('<Button-1>', self.clic) #quand clic gauche de souris appel à la fonction clic
        self.jeu()
        self.fenetre.mainloop()

        

```

# Annexe : présentation du jeu d'Awélé

L'Awélé est un jeu d'origine africaine qui se joue sur un plateau (ou tablier) où chacun des 2 joueurs (Sud et Nord) possède 6 cases.  Les règles de ce jeu  sont simples et facilement implémentables.

Voir la page wikipédia (http://fr.wikipedia.org/wiki/Awalé) qui détaille les règles que nous utilisons ici.

### Règles du jeu
Dans la position de départ, toutes les cases sont remplies avec 4 graines.  Le joueur dont
c'est le tour de jeu (on considère que le camp Sud débute toujours la partie.) choisit une de ses cases contenant des graines, et en retire toutes les graines qu'elle contient.  Il sème alors ces graines une par une dans les cases suivantes dans le sens inverse des aiguilles d'une montre.  Au cours de cette pose, s'il repasse sur la case qui contenait les graines au départ, il n'y dépose pas de graine.

Si la dernière graine est semée dans une case de l'adversaire contenant 2 ou 3 graines après la pose, les graines dans cette case sont capturées par le joueur et elles sont alors retirées du jeu (elles sont mises dans le gain du joueur).  Dans ce cas, si l'avant-dernière case est aussi une case ennemie contenant 2 ou 3 graines, les graines en sont aussi capturées, et ainsi de suite tant que des prises sont possibles (toujours dans le camp ennemi).

Il est à noter qu'un coup n'effectue pas forcément une prise, mais si une prise existe à l'issue d'un coup, elle doit obligatoirement être réalisée totalement.

A l'issue du coup ainsi exécuté, il doit rester au moins une graine dans l'une des cases de l'adversaire (il ne faut pas ''affamer'' l'adversaire) sinon la position est considérée comme illégale et le coup ne peut pas être joué.

Dès qu'un joueur a capturé 25 graines ou plus, il est déclaré vainqueur de la partie et le jeu s'arrête.

Dans le cas où un joueur ne peut pas jouer, la partie s'arrête et toutes les graines restant sur le tablier sont capturées par son adversaire.  Dans ce cas, le joueur qui a capturé le plus de graines a gagné.

### Représentation du tablier

Le tablier est toujours représenté avec les cases de Sud en bas.  
Les cases sont ordonnées et repérées par des chiffres, de 1 à 6, de la gauche vers la droite. Dans chaque case, on indique le nombre de graines qu'elle contient. Ainsi, le tablier de départ classique est donné ci-dessous:

Un coup est noté en donnant le camp qui joue ainsi que le numéro de la case qui est vidée.
